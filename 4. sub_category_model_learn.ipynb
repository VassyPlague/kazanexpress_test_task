# Importing all Data

```python
import pandas as pd
cat_0 = pd.read_parquet('big_categorical_parsed_data/cat_0.parquet').rename(columns={'cat_2' : 'category'}).drop(columns = ['cat_1']).reset_index(drop = True)
cat_1 = pd.read_parquet('big_categorical_parsed_data/cat_1.parquet').rename(columns={'cat_2' : 'category'}).drop(columns = ['cat_1']).reset_index(drop = True)
cat_2 = pd.read_parquet('big_categorical_parsed_data/cat_2.parquet').rename(columns={'cat_2' : 'category'}).drop(columns = ['cat_1']).reset_index(drop = True)
cat_3 = pd.read_parquet('big_categorical_parsed_data/cat_3.parquet').rename(columns={'cat_2' : 'category'}).drop(columns = ['cat_1']).reset_index(drop = True)
cat_4 = pd.read_parquet('big_categorical_parsed_data/cat_4.parquet').rename(columns={'cat_2' : 'category'}).drop(columns = ['cat_1']).reset_index(drop = True)
```

```python
cat_0
```

```python
cat_0[cat_0['category_id'] == 12171]
```

```python
def to_lower(text):
    text = text.split()
    output = ''
    for i in text:
        output += i.lower() + ' '
    return output
```

```python
cat_0.title = cat_0.title.apply(func = lambda x: to_lower(x))
cat_1.title = cat_1.title.apply(func = lambda x: to_lower(x))
cat_2.title = cat_2.title.apply(func = lambda x: to_lower(x))
cat_3.title = cat_3.title.apply(func = lambda x: to_lower(x))
cat_4.title = cat_4.title.apply(func = lambda x: to_lower(x))
```

```python
for i, text in enumerate(cat_0.description):
    if len(str(text)) == 0:
        cat_0.description.iloc[i] = cat_0.title.iloc[i]
        
for i, text in enumerate(cat_1.description):
    if len(str(text)) == 0:
        cat_1.description.iloc[i] = cat_1.title.iloc[i]

for i, text in enumerate(cat_2.description):
    if len(str(text)) == 0:
        cat_2.description.iloc[i] = cat_2.title.iloc[i]
        
for i, text in enumerate(cat_3.description):
    if len(str(text)) == 0:
        cat_3.description.iloc[i] = cat_3.title.iloc[i]
        
for i, text in enumerate(cat_4.description):
    if len(str(text)) == 0:
        cat_4.description.iloc[i] = cat_4.title.iloc[i]
        
```

# 0 Category

```python
cat_0['category'], _ = cat_0['category'].factorize()
```

```python
from sklearn.model_selection import train_test_split
train_cat_0, test_cat_0 = train_test_split (cat_0.drop(columns=['category_id']), test_size= 0.2, random_state= 14)
```

```python
from catboost import CatBoostClassifier

text_data = ['description', 'title']

model_catboost_cat_0 = CatBoostClassifier(
                           task_type="GPU",
                           devices='0:1',
                           text_features= text_data,
                           depth = 8,
                           iterations = 500)
model_catboost_cat_0.fit(train_cat_0.drop(columns = ['category']), train_cat_0['category'])

predicted_catboost_cat_0 = model_catboost_cat_0.predict(test_cat_0.drop(columns = ['category']))
```

```python
model_catboost_cat_0.save_model('models/model_catboost_cat_0_sub',
           format="cbm",
           export_parameters=None,
           pool=None)
```
